```python
from kafka import KafkaConsumer

try:
    consumer = KafkaConsumer(
        'tracking.web.events',
        bootstrap_servers=[
            'localhost:29092',
            'localhost:39092',
            'localhost:49092'
        ],
        group_id='my-group',
        auto_offset_reset='earliest',
        enable_auto_commit=True,
        value_deserializer=lambda v: v.decode('utf-8')
    )

    for msg in consumer:
        print(msg.value)
except KeyboardInterrupt:
    print("Stopping consumer...")
finally:
    consumer.close()
```
